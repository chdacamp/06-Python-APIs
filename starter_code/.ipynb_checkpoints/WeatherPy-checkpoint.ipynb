```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
print("Beginning Data Retrieval")

count = 0
weather_df = pd.DataFrame()

for city in cities:
    count += 1
    try:
        request = requests.get('http://api.openweathermap.org/data/2.5/weather?q='+city+'&APPID='+api_key).json()
        Cloudiness.append(request['clouds']['all'])
        Country.append(request['sys']['country'])
        Date.append(request['dt'])
        Humidity.append(request['main']['humidity'])
        Lat.append(request['coord']['lat'])
        Lng.append(request['coord']['lon'])
        Max_Temp.append(request['main']['temp_max'])
        Wind_Speed.append(request['wind']['speed'])
        City.append(city)
        
        print("Processing Record "+str(count)+" | "+city)
    except:
        Cloudiness.append('NaN')
        Country.append('NaN')
        Date.append('NaN')
        Humidity.append('NaN')
        Lat.append('NaN')
        Lng.append('NaN')
        Max_Temp.append('NaN')
        Wind_Speed.append('NaN')
        City.append('NaN')
        print("City Not Found - SKIP")
        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
print(len(City))
print(len(Cloudiness))
print(len(Country))

weather_dict = {
    "City":City,
    "Cloudiness":Cloudiness,
    "Country":Country,
    "Date":Date,
    "Humidity":Humidity,
    "Lat":Lat,
    "Lng":Lng,
    "Max_Temp":Max_Temp,
    "Wind_Speed":Wind_Speed
}

weather_df = pd.DataFrame(weather_dict)
weather_df = weather_df.set_index("City")
weather_df.drop("NaN",inplace = True)

#convert kelvin to F

weather_df["Max_Temp"] = (weather_df["Max_Temp"] - 273.15) * 9/5 + 32

weather_df
```

```python
weather_df.to_csv("./weather_df.csv")
```

## 3 Observable Trends (In November 2018)

1) Max Temperature is lowest at higher latitudes, relatively flat at low/mid latitudes, with a slight dip at low latitudes

2) Cloudiness does not correlate with latitude

3) Low and high latitudes are more likely to have higher wind speeds
